```python
import numpy as np
```

```python
def TSQR(A):
    m,n = A.shape
    A1 = A[0,m//4,:]
    A_blocks = [A[(m*k)//4,m*(k+1)//4 :] for k in range(4)] 
    
    Q_blocks = []
    R_blocks = []

    for Ai in A_blocks:
        Qi, Ri = np.linalg.qr(Ai)
        Q_blocks.append(Qi)
        R_blocks.append(Ri)
        
    R_stacked = np.vstack(R_blocks)
    Q2, R_final = np.linalg.qr(R_stacked, mode='reduced')
    Q2_blocks = np.array_split(Q2, 4)

    Q_final_blocks = []
    for Qi, Q2i in zip(Q_blocks, Q2_blocks):
        Q_final_blocks.append(Qi @ Q2i)

    Q_final = np.vstack(Q_final_blocks)

    return Q_final, R_final
```
